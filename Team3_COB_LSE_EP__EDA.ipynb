```python
# Import Libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc = {'figure.figsize':(15,10)})
```

```python
# Load the data
```

```python
central = pd.read_csv('c_london.csv')
```

```python
central.head(10)
```

```python
central.shape
```

```python
# check for data types
```

```python
central.dtypes
```

```python
# checking for missing values
```

```python
central.isnull().sum()
```

```python
central['Borough'].unique()
```

```python
# Cycle Traffic by Borough
```

```python
c_borough=central.groupby('Borough')[['Total cycles']].sum().sort_values('Total cycles', ascending=False)
```

```python
c_borough
```

```python
c_borough.reset_index(inplace=True)
```

```python
c_borough
```

```python
# Cycle Traffic by Borough
```

```python
sns.barplot(x='Total cycles',y='Borough',data=c_borough)
plt.show()
```

```python
sns.barplot(x="Total cycles", y="Borough", hue="survey day",estimator=sum,data=central)
plt.show()
```

```python
# Cycle Traffic by Day
```

```python
c_day=central.groupby('survey day')[['Total cycles']].sum().sort_values('Total cycles', ascending=False)
```

```python
c_day
```

```python
c_day.reset_index(inplace=True)
```

```python
sns.barplot(x='Total cycles',y='survey day',data=c_day)
plt.show()
```

```python
# Cycle Traffic by month
```

```python
c_month=central.groupby('survey month')[['Total cycles']].sum().sort_values('Total cycles', ascending=False)
```

```python
c_month
```

```python
c_month.reset_index(inplace=True)
```

```python
sns.barplot(x='Total cycles',y='survey month',data=c_month)
plt.show()
```
